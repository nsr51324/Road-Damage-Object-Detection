```python
from ultralytics import YOLO
import pandas as pd
import matplotlib.pyplot as plt
import os
data_path = "/content/road-damage-1/data.yaml"
epochs = 50
imgsz = 640
batch = 16
```

**YOLO8**

```python
model_v8 = YOLO("yolov8n.pt")
model_v8.train(
    data=data_path,
    epochs=50,
    imgsz=640,
    batch=16,
    patience=10,
    name="yolov8_road"
)
```

**YOLO10**

```python
model_v10 = YOLO("yolov10n.pt")
model_v10.train(
    data=data_path,
    epochs=50,
    imgsz=640,
    batch=16,
    patience=3,
    name="yolov10_road"
)
```

**YOLO11**

```python
model_v11 = YOLO("yolo11n.pt")
model_v11.train(
    data=data_path,
    epochs=50,
    imgsz=640,
    batch=16,
    patience=3,
    name="yolov11_road"
)
```

Evaluation Metrics

```python
metrics_v8 = model_v8.val()
metrics_v10 = model_v10.val()
metrics_v11 = model_v11.val()
```

```python
def get_size(path):
    return round(os.path.getsize(path) / (1024 * 1024), 2)
size_v8 = get_size("/content/runs/detect/yolov8_road/weights/best.pt")
size_v10 = get_size("/content/runs/detect/yolov10_road/weights/best.pt")
size_v11 = get_size("/content/runs/detect/yolov11_road-2/weights/best.pt")
```

```python
comparison = pd.DataFrame({
    "Model": ["YOLOv8", "YOLOv10", "YOLO11"],
    "mAP50": [metrics_v8.box.map50, metrics_v10.box.map50, metrics_v11.box.map50],
    "mAP50-95": [metrics_v8.box.map, metrics_v10.box.map, metrics_v11.box.map],
    "Precision": [metrics_v8.box.mp, metrics_v10.box.mp, metrics_v11.box.mp],
    "Recall": [metrics_v8.box.mr, metrics_v10.box.mr, metrics_v11.box.mr],
    "Model_Size_MB": [size_v8, size_v10, size_v11],
    "FPS": [1000 / metrics_v8.speed["inference"], 1000 / metrics_v10.speed["inference"], 1000 / metrics_v11.speed["inference"]]
})
comparison
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 4))
ax[0].bar(comparison["Model"], comparison["mAP50-95"], color=["blue", "green", "orange"])
ax[0].set_title("mAP50-95")
ax[1].bar(comparison["Model"], comparison["FPS"], color=["blue", "green", "orange"])
ax[1].set_title("FPS")
ax[2].bar(comparison["Model"], comparison["Model_Size_MB"], color=["blue", "green", "orange"])
ax[2].set_title("Model Size MB")
plt.tight_layout()
plt.show()
```
